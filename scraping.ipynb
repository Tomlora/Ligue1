# Equipe de Bordeaux plusieurs années

```python
url_home = "https://fbref.com/fr/comps/13/Statistiques-Ligue-1"
```

```python
years = list(range(2022, 2020, -1))
all_matches = []

for year in years:
    # même procédé qu'au début du notebook
    # on récupère la page
    data = requests.get(url_home)
    # on lit le code html... 
    soup = BeautifulSoup(data.text)
    # <table class="stats_table sortable min_width now_sortable" id="results111831_overall" data-cols-to-freeze=",2"> inspecter la page
    # [0] car on veut la première table qui résume l'ensemble.. [1] par exemple, est le tableau domicile/extérieur
    table = soup.select('table.stats_table')[0]
    
    # on cherche les tags avec un "a" (table.find_all('a'))
    # Dans les liens recueillis dans links, on prend ce qui contient "href"
    links = [l.get('href') for l in table.find_all('a')]
    # on prend les links qui contiennent '/equipes/'
    links = [l for l in links if '/equipes/' in l]
    # on rajoute la source de base aux urls d'équipes
    team_urls = [f'https://fbref.com{l}' for l in links] 
    
    # permet d'identifier l'url de l'année précédente... Ca va servir à maj url_home pour la boucle
    previous_season = soup.select('a.prev')[0].get("href")
    url_home = f"https://fbref.com{previous_season}" 
    
    #tqdm pour suivre l'avancement
    for team_url in tqdm(team_urls):
        #permet de découper l'url pour n'avoir que le nom de l'équipe.
        team_name = team_url.split("/")[-1].replace("Statistiques", "").replace("-", "") 
        
        # on lit le code html
        data = requests.get(team_url)
        # on veut le tableau "Calendrier et résultats". Les balises des tableaux ont <caption>
        matches = pd.read_html(data.text, match = "Calendrier et résultats ")
        
        # on refait la même chose pour les tirs
        
        # on lit le code html... 
        soup = BeautifulSoup(data.text)
        # on cherche les tags avec un "a" dans le code html entier
        links = soup.find_all('a')
        # Dans les liens recueillis dans links, on prend ce qui contient "href"
        links = [l.get("href") for l in links]
         # on prend les links qui contiennent 'all_comps/shooting'
        links = [l for l in links if l and 'all_comps/shooting' in l]
        # on rajoute la source de base aux urls de tirs
        tirs_equipe = requests.get(f'https://fbref.com{links[0]}')
        # on veut le tableau "Tirs ". Les balises des tableaux ont <caption>
        tirs_equipe = pd.read_html(tirs_equipe.text, match="Tirs ")[0]
        # on enlève le niveau en trop, comme avant...
        tirs_equipe.columns = tirs_equipe.columns.droplevel()
        
        # on met tout dans un dataframe
        try:
            team_data = matches[0].merge(tirs_equipe[['Date', 'Dist', 'CF', 'PénM', 'PénT', 'Buts', 'Tirs', 'TC', 'TC%', 'B/Tir', 'B/TC']], on="Date")
        except ValueError:
            continue
        
        # Feature engineering
        team_data = team_data[team_data["Comp"] == "Ligue 1"]
        team_data["Saison"] = year
        team_data["Equipe"] = team_name
        # on ajoute notre nouveau dataframe à une liste de dataframes
        all_matches.append(team_data)
        time.sleep(1) # éviter que le site nous bloque pour spam
        
        
    
    
```

```python
# Source
links
```

```python
match_df = pd.concat(all_matches)
```
